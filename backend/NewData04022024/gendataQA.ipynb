```python
from llama_index.core import SimpleDirectoryReader
from llama_index.core.node_parser import SimpleNodeParser
from llama_index.core.schema import MetadataMode

def load_corpus(directory, verbose=False):
    if verbose:
        print(f"Loading files in {directory}")

    required_exts = [".pdf", ".docx"]
    # required_exts = [".png"]
    reader = SimpleDirectoryReader(directory, required_exts=required_exts)
    docs = reader.load_data()
    print(docs)
    if verbose:
        print(f"Loaded {len(docs)} docs")

    # parser = SimpleNodeParser.from_defaults()
    parser = SentenceSplitter(chunk_size=2000, chunk_overlap=150)
    nodes = parser.get_nodes_from_documents(docs, show_progress=verbose)

    if verbose:
        print(f"Parsed {len(nodes)} nodes")

    return docs
```

```python
train_nodes = load_corpus(TRAIN_FILES, verbose=True)
```

```python
template = (
"""Thông tin ngữ cảnh ở bên dưới.\n
----------------------
{context_str}
------------- ---------\
Với thông tin ngữ cảnh chứ không phải kiến thức có sẵn.
chỉ tạo câu hỏi dựa trên truy vấn bên dưới.
Bạn là Giáo viên/Giáo sư. 
Nhiệm vụ của bạn là thiết lập {num_questions_per_chunk} câu hỏi cho bài kiểm tra/bài kiểm tra sắp tới.
Các câu hỏi nên đa dạng về bản chất trong toàn bộ tài liệu. 
Giới hạn các câu hỏi trong phạm vi thông tin ngữ cảnh được cung cấp.
"""
)
```

```python
llm = OpenAI(model="gpt-3.5-turbo", temperature=0)

dataset_generator = DatasetGenerator.from_documents(
    documents=train_nodes,
    llm=llm,
    num_questions_per_chunk=10,  # set the number of questions per nodes
    text_qa_template=template,
)
```

```python
rag_dataset = dataset_generator.generate_questions_from_nodes()
```
