```python
""" 
CODE: Mã ngành xét tuyển	
NAME: Tên ngành
Xét tuyển tài năng: Số chỉ tiêu xét tuyển tài năng
THPT: Số chỉ tiêu thi tốt nghiệp THPT
XTKH: Số chỉ tiêu xét tuyển kết hợp
ĐGNL, ĐGTD: Số chỉ tiêu đánh giá năng lực, đánh giá tư duy
CODE,NAME,THPT,XTKH,DGNL
7520207, Kỹ thuật Điện tử viễn thông, 195, 120, 75
=> - Ngành Kỹ thuật Điện tử viễn thông (Mã ngành: 7520207) tổng chỉ tiêu 390 gồm 15 chỉ tiêu xét tuyển tài năng, 195 chỉ tiêu thi THPT, 120 chỉ tiêu xét tuyển kết hợp, 75 chỉ tiêu đánh giá năng lực, đánh giá tư duy.
 """

df_to_dict = df.to_dict(orient='records')

a = []
for item in df_to_dict:
    index = df_to_dict.index(item)
    if index + 1 < 19:
        str = f"{index+1}. chỉ tiêu tuyển sinh ngành {item['NAME']} tại cơ sở phía Bắc (BVH) (Mã ngành {item['CODE']}) là {item['Xét tuyển tài năng'] + item['THPT'] + item['XTKH'] + item['ĐGNL, ĐGTD']} gồm {item['Xét tuyển tài năng']} chỉ tiêu xét tuyển tài năng, {item['THPT']} chỉ tiêu thi THPT, {item['XTKH']} chỉ tiêu xét tuyển kết hợp, {item['ĐGNL, ĐGTD']} chỉ tiêu đánh giá năng lực, đánh giá tư duy."
        a.append([item['CODE'], str])
    else:
        str = f"{index+1}. chỉ tiêu tuyển sinh ngành {item['NAME']} tại cơ sở phía Nam (BVS) (Mã ngành {item['CODE']}) là {item['Xét tuyển tài năng'] + item['THPT'] + item['XTKH'] + item['ĐGNL, ĐGTD']} gồm {item['Xét tuyển tài năng']} chỉ tiêu xét tuyển tài năng, {item['THPT']} chỉ tiêu thi THPT, {item['XTKH']} chỉ tiêu xét tuyển kết hợp, {item['ĐGNL, ĐGTD']} chỉ tiêu đánh giá năng lực, đánh giá tư duy."
        a.append([item['CODE'], str])

a.sort(key=lambda x: x[0])

for item in a:
    print(item[1])
    
```

```python
# Index
from langchain_openai import OpenAIEmbeddings
from langchain_community.vectorstores import FAISS

embed_model = OpenAIEmbeddings()
vectorstore = FAISS.load_local(
    "../NewVectorDB/",
    OpenAIEmbeddings(),
    allow_dangerous_deserialization=True,
)


retriever = vectorstore.as_retriever(search_kwargs={"k": 5, "threshold": 0.5})
retriever
```

```python
# -*- coding: utf-8 -*-

from langchain.text_splitter import RecursiveCharacterTextSplitter, CharacterTextSplitter
from langchain_community.document_loaders import PyPDFLoader, DirectoryLoader, TextLoader
from langchain_community.vectorstores import FAISS
from langchain_community.document_loaders import TextLoader


from torch import cuda
from langchain.embeddings.huggingface import HuggingFaceEmbeddings
from langchain.embeddings import OpenAIEmbeddings
```

```python
loader = TextLoader("./Chỉ tiêu tuyển sinh.txt")

documents = loader.load()

text_splitter = RecursiveCharacterTextSplitter(chunk_size=2048, chunk_overlap=256)
chunks = text_splitter.split_documents(documents)
chunks
```

```python
# Embeding
embed_model = OpenAIEmbeddings()
vectorstore.add_documents(chunks)
```
