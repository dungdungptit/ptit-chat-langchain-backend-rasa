# Docx to subdocx

```python
import os
from docx import Document

def split_docx(input_file):
    # Mở file DOCX đầu vào
    doc = Document(input_file)

    # Tạo thư mục đầu ra nếu chưa tồn tại
    output_dir = "./" + input_file.split('/')[-1].split('.')[0]
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # Lặp qua các đoạn văn bản trong tài liệu
    for i, paragraph in enumerate(doc.paragraphs):
        # Kiểm tra nếu đoạn văn bản có heading (kiểu tiêu đề)
        if paragraph.style.name.startswith('Heading'):
            pre_level_heading = paragraph.style.name.split(' ')[1]
            # Tạo một tài liệu mới cho mỗi heading
            print(paragraph.style.name)
            new_doc = Document(docx=None)

            if i < len(doc.paragraphs) - 1:
                # Sao chép heading và các đoạn văn bản liên quan
                heading_text = paragraph.text.strip().replace(" ", "_").lower()
                new_doc.add_paragraph(paragraph.text, style=paragraph.style)
                j = i + 1
                # while j < len(doc.paragraphs) and not doc.paragraphs[j].style.name.startswith('Heading'):
                while j < len(doc.paragraphs):
                    if doc.paragraphs[j].style.name.startswith('Heading'):
                        level_heading = doc.paragraphs[j].style.name.split(' ')[1]
                        if level_heading <= pre_level_heading:
                            break
                    new_doc.add_paragraph(doc.paragraphs[j].text, style=doc.paragraphs[j].style)
                    j += 1

                # Lưu tài liệu mới thành một file DOCX
                new_doc.save(os.path.join(output_dir, f"{heading_text}.docx"))
                print(f"File '{heading_text}.docx' created.")
                for p in new_doc.paragraphs:
                    print(p.text)

if __name__ == "__main__":
    input_file = "./De an tuyen sinh 2024_Final.docx"
    split_docx(input_file)
    # for file in list_dir:
    #     split_docx('./' + file)
```

# Subdocx to text

```python
from llama_index.core.node_parser import SimpleNodeParser
from llama_index.core.schema import MetadataMode
from llama_index.core.node_parser import SentenceSplitter

TRAIN_FILES = "./"
```

```python
from llama_index.core import SimpleDirectoryReader
from llama_index.core.node_parser import SimpleNodeParser
from llama_index.core.schema import MetadataMode

def load_corpus(directory, verbose=False):
    if verbose:
        print(f"Loading files in {directory}")

    required_exts = [".pdf", ".docx"]
    # required_exts = [".png"]
    reader = SimpleDirectoryReader(directory, required_exts=required_exts)
    docs = reader.load_data()
    if verbose:
        print(f"Loaded {len(docs)} docs")

    # parser = SimpleNodeParser.from_defaults()
    parser = SentenceSplitter(chunk_size=2512, chunk_overlap=256)
    nodes = parser.get_nodes_from_documents(docs, show_progress=verbose)

    if verbose:
        print(f"Parsed {len(nodes)} nodes")

    return nodes
```
